# Filtering with where method

```python
bond = pd.read_csv("csvfiles/jamesbond.csv" , index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask].head()
```

# The query() method
### Works only when column names dont have spaces

```python
bond = pd.read_csv("csvfiles/jamesbond.csv" , index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.columns = [col_name.replace(" " , "_") for col_name in bond.columns]
bond.columns
```

```python
bond.query('Actor == "Sean Connery"').head()
# people have mentioned that its faster
```

```python
bond.query('Actor == "Roger Moore" and Director == "John Glen"')
```

```python
bond.query('Actor in ["Roger Moore" , "Sean Connery"]')
```

# .apply()

```python
bond = pd.read_csv("csvfiles/jamesbond.csv" , index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def convert_and_add_millions(number):
    return str(number) + " Millions!"

bond["Bond Actor Salary"] = bond["Bond Actor Salary"].apply(convert_and_add_millions)
    
```

```python
bond.head()
```

```python
bond = pd.read_csv("csvfiles/jamesbond.csv" , index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Roger Moore":
        return "The best!"
    elif actor == "Sean Connery" and budget > 50:
        return "Awesome!"
    else:
        return "Not sure!"
```

```python
bond["Review"] = bond.apply(good_movie , axis="columns")
```
