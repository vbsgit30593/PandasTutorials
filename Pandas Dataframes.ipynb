# Rank values

```python
nba = pd.read_csv("csvfiles/nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Rank"] = nba["Salary"].rank(ascending=False).astype(int)
```

```python
nba.sort_values("Rank" , ascending="False").head()
```
